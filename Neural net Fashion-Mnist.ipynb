## Question 1

     Download Fashion-Mnist Dataset and plot one image of each class

```python
from keras.datasets import fashion_mnist
```

```python
(train_data,train_label),(test_data,test_label) = fashion_mnist.load_data()
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", 
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
each_class_image_index = []
classes = []
for i in range(len(train_label)):
    if train_label[i] not in classes:
        each_class_image_index.append(i)
        classes.append(train_label[i])
    if len(each_class_image_index) == 10:
        break
```

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 5))

for ax, idx in zip(axes.ravel(), each_class_image_index):
    ax.imshow(train_data[idx], cmap='gray')
    ax.set_title(f"Class {class_names[train_label[idx]]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
