```python
class Neural_net:
    def __init__(self,num_hidden_layers,num_neurons_each_layer,input_size):
        self.num_hidden_layers = num_hidden_layers
        self.num_neurons_each_layer = num_neurons_each_layer
        self.input_size = input_size
        self.W,self.B = self.init_parameters()

    def init_parameters(self):
        W = []
        B = []
        W.append(np.random.randn(self.num_neurons_each_layer[0], self.input_size) * np.sqrt(1 / self.input_size))
        for i in range(1,self.num_hidden_layers):
            W.append(np.random.randn(self.num_neurons_each_layer[i], self.num_neurons_each_layer[i-1]) * np.sqrt(1 / self.num_neurons_each_layer[i-1]))

        B.append(np.zeros(shape=(1,self.num_neurons_each_layer[0])))
        for i in range(1,self.num_hidden_layers):
            B.append(np.zeros(shape=(1,self.num_neurons_each_layer[i])))

        return W, B

    def softmax_activation(self,x):
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

    def ReLU(self,x):
        return np.maximum(0,x)

    def ReLU_derivtive(self,x):
        return np.where(x > 0, 1, 0)
    
    def sigmoid_activation(self,x):
        x = np.clip(x, -500, 500)
        return 1/(1+np.exp(-x))

    def sigmoid_derivative(self,x):
        return self.sigmoid_activation(x)*(1-self.sigmoid_activation(x))

    def FeedForward(self,data):
        a_l = []
        h_l = []
        input_linear_out = np.dot(data,self.W[0].T) + self.B[0]
        a_l.append(input_linear_out)
        for i in range(1,self.num_hidden_layers):
            activation_out = self.ReLU(a_l[-1])
            h_l.append(activation_out)
            linear_out = np.dot(h_l[-1],self.W[i].T) + self.B[i]
            a_l.append(linear_out)


        y_hat = self.softmax_activation(a_l[-1])
        return a_l,h_l,y_hat

    def val_loss_and_acc(self,val_data,val_label):
        acc = 0
        error = 0
        val_loss = []
        val_acc = []
        
        for d in range(val_data.shape[0]):
            a_li,h_li,y_hat_i = self.FeedForward(val_data[d].reshape(1,-1))
            error += -np.sum(val_label[d] * np.log(y_hat_i[0]))
            if np.argmax(val_label[d]) == np.argmax(y_hat_i[0]):
              acc += 1
        return error/val_data.shape[0],acc/val_data.shape[0]

    def stochastic_gradient_descent(self,epochs,learning_rate,data,label,val_data,val_label):
        for ep in range(epochs):
            train_loss = 0
            train_acc = 0
            for d in range(len(data)):
              a_li,h_li,y_hat_i = self.FeedForward(data[d].reshape(1,-1))
              train_loss += -np.sum(label[d] * np.log(y_hat_i[0]))
                
              dw,db = self.backpropagation(a_li,h_li,y_hat_i[0],label[d],data[d])
              if np.argmax(train_label[d]) == np.argmax(y_hat_i[0]):
                  train_acc += 1
                  
              for i in range(self.num_hidden_layers):
                self.W[i] -= (1/data.shape[0])*learning_rate * dw[i]
                self.B[i] -= learning_rate * db[i]


                if((d+1)%1000 == 0):
                    
                    
                    
                    val_loss,val_acc = self.val_loss_and_acc(val_data,val_label)
                    wandb.log({'epoch':ep+1,'train_loss':train_loss/(d+1),'train_acc':train_acc/(d+1),'val_loss':val_loss,'val_acc':val_acc})
                    sys.stdout.write(f"\rEpoch {ep+1}/{epochs} - {d+1}/{len(data)} - Loss: {train_loss/(d+1):.6f} Train-Acc: {train_acc/(d+1):.6f} val-loss:{val_loss:.6f} val-Acc: {val_acc:.6f} ")
                    sys.stdout.flush()
            print()
            
    def momentum_gradient_descent(self, epochs, learning_rate, data, label, val_data, val_label, beta=0.6):

        momentum_W = [np.zeros_like(w) for w in self.W]
        momentum_B = [np.zeros_like(b) for b in self.B]
    
        for ep in range(epochs):
            train_loss = 0
            train_acc = 0
            for d in range(len(data)):
                a_li, h_li, y_hat_i = self.FeedForward(data[d].reshape(1, -1))
                train_loss += -np.sum(label[d] * np.log(y_hat_i[0]))
                if np.argmax(label[d]) == np.argmax(y_hat_i[0]):
                    train_acc += 1
    
                dw, db = self.backpropagation(a_li, h_li, y_hat_i[0], label[d], data[d])
    
                # Update momentum terms
                for i in range(self.num_hidden_layers):
                    momentum_W[i] = beta * momentum_W[i] + dw[i]
                    momentum_B[i] = beta * momentum_B[i] + db[i]
    
                    # Update weights using momentum
                    self.W[i] -= learning_rate*momentum_W[i]
                    self.B[i] -= learning_rate*momentum_B[i]
    
                if ((d + 1) % 1000 == 0):
                    val_loss, val_acc = self.val_loss_and_acc(val_data, val_label)
                    wandb.log({
                        'epoch': ep + 1,
                        'train_loss': train_loss / (d + 1),
                        'train_acc': train_acc / (d + 1),
                        'val_loss': val_loss,
                        'val_acc': val_acc
                    })
                    sys.stdout.write(f"\rEpoch {ep + 1}/{epochs} - {d + 1}/{len(data)} - Loss: {train_loss / (d + 1):.6f} Train-Acc: {train_acc / (d + 1):.6f} val-loss:{val_loss:.6f} val-Acc: {val_acc:.6f} ")
                    sys.stdout.flush()
            print()

    def nesterov_accelerated_gradient_descent(self,epochs, learning_rate, data, label, val_data, val_label, beta=0.6):
        momentum_W = [np.zeros_like(w) for w in self.W]
        momentum_B = [np.zeros_like(b) for b in self.B]
    
        for ep in range(epochs):
            train_loss = 0
            train_acc = 0
            for d in range(len(data)):
                
                W_lookahead = [w - beta * v for w, v in zip(self.W, momentum_W)]
                B_lookahead = [b - beta * v for b, v in zip(self.B, momentum_B)]

                w_original = self.W
                b_original = self.B
                self.W, self.B = W_lookahead, B_lookahead
                
                a_li, h_li, y_hat_i = self.FeedForward(data[d].reshape(1, -1))
                train_loss += -np.sum(label[d] * np.log(y_hat_i[0]))
                if np.argmax(label[d]) == np.argmax(y_hat_i[0]):
                    train_acc += 1
    
                # Compute gradients at the look-ahead position
                dw, db = self.backpropagation(a_li, h_li, y_hat_i[0], label[d], data[d])

                self.W = w_original
                self.B = b_original
                
                # Update velocity and weights
                for i in range(self.num_hidden_layers):
                    momentum_W[i] = beta * momentum_W[i] + dw[i]
                    momentum_B[i] = beta * momentum_B[i] + db[i]


                    self.W[i] -= learning_rate * momentum_W[i]
                    self.B[i] -= learning_rate * momentum_B[i]
    
                if ((d + 1) % 1000 == 0):
                    val_loss, val_acc = self.val_loss_and_acc(val_data, val_label)
                    wandb.log({
                        'epoch': ep + 1,
                        'train_loss': train_loss / (d + 1),
                        'train_acc': train_acc / (d + 1),
                        'val_loss': val_loss,
                        'val_acc': val_acc
                    })
                    sys.stdout.write(f"\rEpoch {ep + 1}/{epochs} - {d + 1}/{len(data)} - Loss: {train_loss / (d + 1):.6f} Train-Acc: {train_acc / (d + 1):.6f} val-loss:{val_loss:.6f} val-Acc: {val_acc:.6f} ")
                    sys.stdout.flush()
            print()

    def RMSprop(self,epochs,learning_rate,data,label,val_data,val_label,beta=0.2,eps=1e-4):
        momentum_W = [np.zeros_like(w) for w in self.W]
        momentum_B = [np.zeros_like(b) for b in self.B]
    
        for ep in range(epochs):
            train_loss = 0
            train_acc = 0
            for d in range(len(data)):
                a_li, h_li, y_hat_i = self.FeedForward(data[d].reshape(1, -1))
                train_loss += -np.sum(label[d] * np.log(y_hat_i[0]))
                if np.argmax(label[d]) == np.argmax(y_hat_i[0]):
                    train_acc += 1
    
                dw, db = self.backpropagation(a_li, h_li, y_hat_i[0], label[d], data[d])
    
                # Update momentum terms
                for i in range(self.num_hidden_layers):
                    momentum_W[i] = beta * momentum_W[i] + (1-beta)*(dw[i]**2)
                    momentum_B[i] = beta * momentum_B[i] + (1-beta)*(db[i]**2)
    
                    # Update weights using momentum
                    adaptive_lr_W = learning_rate/np.sqrt(momentum_W[i]+eps)
                    adaptive_lr_B = learning_rate/np.sqrt(momentum_B[i]+eps)
                    self.W[i] -= adaptive_lr_W*dw[i]
                    self.B[i] -= adaptive_lr_B*db[i]
    
                if ((d + 1) % 1000 == 0):
                    val_loss, val_acc = self.val_loss_and_acc(val_data, val_label)
                    wandb.log({
                        'epoch': ep + 1,
                        'train_loss': train_loss / (d + 1),
                        'train_acc': train_acc / (d + 1),
                        'val_loss': val_loss,
                        'val_acc': val_acc
                    })
                    sys.stdout.write(f"\rEpoch {ep + 1}/{epochs} - {d + 1}/{len(data)} - Loss: {train_loss / (d + 1):.6f} Train-Acc: {train_acc / (d + 1):.6f} val-loss:{val_loss:.6f} val-Acc: {val_acc:.6f} ")
                    sys.stdout.flush()
            print()

    def Adam(self,epochs,learning_rate,data,label,val_data,val_label,beta1=0.9,beta2=0.999,eps=1e-4):
        momentum1_W = [np.zeros_like(w) for w in self.W]
        momentum1_B = [np.zeros_like(b) for b in self.B]
        momentum2_W = [np.zeros_like(w) for w in self.W]
        momentum2_B = [np.zeros_like(b) for b in self.B]

        
        t = 0
    
        for ep in range(epochs):
            train_loss = 0
            train_acc = 0
            for d in range(len(data)):
                t += 1
                a_li, h_li, y_hat_i = self.FeedForward(data[d].reshape(1, -1))
                train_loss += -np.sum(label[d] * np.log(y_hat_i[0]))
                if np.argmax(label[d]) == np.argmax(y_hat_i[0]):
                    train_acc += 1
    
                dw, db = self.backpropagation(a_li, h_li, y_hat_i[0], label[d], data[d])
    
                # Update momentum terms
                for i in range(self.num_hidden_layers):
                    momentum1_W[i] = beta1 * momentum1_W[i] + (1-beta1)*(dw[i])
                    momentum1_B[i] = beta1 * momentum1_B[i] + (1-beta1)*(db[i])

                    momentum2_W[i] = beta2 * momentum2_W[i] + (1-beta2)*(dw[i]**2)
                    momentum2_B[i] = beta2 * momentum2_B[i] + (1-beta2)*(db[i]**2)
                    
                    momentum1_W_hat = momentum1_W[i]/(1-(beta1**t))
                    momentum1_B_hat = momentum1_B[i]/(1-(beta1**t))
                    
                    momentum2_W_hat = momentum2_W[i]/(1-(beta2**t))
                    momentum2_B_hat = momentum2_B[i]/(1-(beta2**t))
    
                    # Update weights using momentum
                    adaptive_lr_W = (learning_rate/(np.sqrt(momentum2_W_hat)+eps))
                    adaptive_lr_B = (learning_rate/(np.sqrt(momentum2_B_hat)+eps))
                    self.W[i] -= adaptive_lr_W*momentum1_W_hat
                    self.B[i] -= adaptive_lr_B*momentum1_B_hat
    
                if ((d + 1) % 1000 == 0):
                    val_loss, val_acc = self.val_loss_and_acc(val_data, val_label)
                    wandb.log({
                        'epoch': ep + 1,
                        'train_loss': train_loss / (d + 1),
                        'train_acc': train_acc / (d + 1),
                        'val_loss': val_loss,
                        'val_acc': val_acc
                    })
                    sys.stdout.write(f"\rEpoch {ep + 1}/{epochs} - {d + 1}/{len(data)} - Loss: {train_loss / (d + 1):.6f} Train-Acc: {train_acc / (d + 1):.6f} val-loss:{val_loss:.6f} val-Acc: {val_acc:.6f} ")
                    sys.stdout.flush()
            print()
            
    def NAdam(self,epochs,learning_rate,data,label,val_data,val_label,beta1=0.9,beta2=0.999,eps=1e-8):
        momentum1_W = [np.zeros_like(w) for w in self.W]
        momentum1_B = [np.zeros_like(b) for b in self.B]
        momentum2_W = [np.zeros_like(w) for w in self.W]
        momentum2_B = [np.zeros_like(b) for b in self.B]
        t = 0
        for ep in range(epochs):
            train_loss = 0
            train_acc = 0
            for d in range(len(data)):
                t += 1
                a_li, h_li, y_hat_i = self.FeedForward(data[d].reshape(1, -1))
                train_loss += -np.sum(label[d] * np.log(y_hat_i[0]))
                if np.argmax(label[d]) == np.argmax(y_hat_i[0]):
                    train_acc += 1
    
                dw, db = self.backpropagation(a_li, h_li, y_hat_i[0], label[d], data[d])
    
                # Update momentum terms
                for i in range(self.num_hidden_layers):
                    momentum1_W[i] = beta1 * momentum1_W[i] + (1-beta1)*(dw[i])
                    momentum1_B[i] = beta1 * momentum1_B[i] + (1-beta1)*(db[i])

                    momentum2_W[i] = beta2 * momentum2_W[i] + (1-beta2)*(dw[i]**2)
                    momentum2_B[i] = beta2 * momentum2_B[i] + (1-beta2)*(db[i]**2)
                    
                    momentum1_W_hat = momentum1_W[i]/(1-(beta1**t))
                    momentum1_B_hat = momentum1_B[i]/(1-(beta1**t))
                    
                    momentum2_W_hat = momentum2_W[i]/(1-(beta2**t))
                    momentum2_B_hat = momentum2_B[i]/(1-(beta2**t))
    
                    # Update weights using momentum
                    m_nestrov_W = beta1*momentum1_W_hat + ((1-beta1)*dw[i])/(1-beta1**t)
                    m_nestrov_B = beta1*momentum1_B_hat + ((1-beta1)*db[i])/(1-beta1**t)
                    
                    adaptive_lr_W = (learning_rate/(np.sqrt(momentum2_W_hat)+eps))
                    adaptive_lr_B = (learning_rate/(np.sqrt(momentum2_B_hat)+eps))
                    self.W[i] -= adaptive_lr_W*m_nestrov_W
                    self.B[i] -= adaptive_lr_B*m_nestrov_B
    
                if ((d + 1) % 1000 == 0):
                    val_loss, val_acc = self.val_loss_and_acc(val_data, val_label)
                    wandb.log({
                        'epoch': ep + 1,
                        'train_loss': train_loss / (d + 1),
                        'train_acc': train_acc / (d + 1),
                        'val_loss': val_loss,
                        'val_acc': val_acc
                    })
                    sys.stdout.write(f"\rEpoch {ep + 1}/{epochs} - {d + 1}/{len(data)} - Loss: {train_loss / (d + 1):.6f} Train-Acc: {train_acc / (d + 1):.6f} val-loss:{val_loss:.6f} val-Acc: {val_acc:.6f} ")
                    sys.stdout.flush()
            print()

    def backpropagation(self,A,H,y_hat,y,data):
      dA = []
      dH = []
      dW = []
      dB = []
      dA.append(-(y - y_hat))
      for i in range(self.num_hidden_layers-1,-1,-1):
        if i == self.num_hidden_layers - 1:
            grad_val_w = np.dot(dA[-1].reshape(-1, 1), H[-1].reshape(1, -1))
        elif i == 0:
          grad_val_w = np.dot(dA[-1].reshape(-1,1),data.reshape(1,-1))
        else:
          grad_val_w = np.dot(dA[-1].reshape(-1,1),H[i-1].reshape(1,-1))
        dW.append(grad_val_w)
        grad_val_b = np.sum(dA[-1],axis=0,keepdims=True)
        dB.append(grad_val_b)

        if i > 0:
          grad_L_hi = np.dot(self.W[i].T,dA[-1].T)
          grad_L_ai = np.multiply(grad_L_hi.T,self.ReLU_derivtive(A[i-1]))
          dA.append(grad_L_ai)
      return dW[::-1],dB[::-1]
```
